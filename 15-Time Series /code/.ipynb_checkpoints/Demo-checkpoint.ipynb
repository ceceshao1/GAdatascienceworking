## DEMO: Exploring Rossmann Drug Store Sales Data

- `Store` - a unique ID for each store
- `Sales` - the sales or turnover for any given day (this is what you are predicting)
- `Customers` - the number of customers on a given day
- `Open` - an indicator for whether the store was open: 0 = closed, 1 = open
- `StateHoliday` - indicates a state holiday. Normally all stores, with few exceptions, are closed on state holidays. Note that all schools are closed on public holidays and weekends. a = public holiday, b = Easter holiday, c = Christmas, 0 = None
- `SchoolHoliday` - indicates if the (Store, Date) was affected by the closure of public schools
- `Promo` - indicates whether a store is running a promo on that day

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('../assets/dataset/rossmann.csv', skipinitialspace=True, low_memory=False)

# .to_datetime() converts strings of the Date column to datetime objects.
data['Date'] = pd.to_datetime(data['Date'])

# Set the index to this new date column
data.set_index('Date', inplace=True)

# Create new columns for the month and year from the new datetime objects
data['Year'] = data.index.year
data['Month'] = data.index.month

# Filter the Store data down to the store referenced as 1
store1_data = data[data.Store == 1]

data.head()
```

```python
data.index.year
```

```python
data[['Sales']].resample('M').mean().sort_values(by='Sales')
```

```python
store1_data.index
```

```python
store1_data_2015 = store1_data['2015']
store1_data_2015[store1_data_2015.Open==1][['Sales']].plot();
```

```python
sns.factorplot(
    col='Open',
    hue='Promo',
    x='Month',
    y='Sales',
    data=store1_data, 
    kind='box');
```

```python
sns.factorplot(
    col='Open',
    x='DayOfWeek',
    y='Customers',
    data=store1_data,
    kind='box',);
```

```python
data[['Sales']].resample('D').mean().rolling(window=15, center=False).mean().diff(1).sort_values(by='Sales')
```

```python
average_daily_customers = data[['Customers']].resample('D').mean()
average_daily_sales = data[['Sales', 'Open']].resample('D').mean()
```

```python
average_daily_customers['DiffVsLastWeek'] = average_daily_customers.diff(periods=7)
average_daily_sales['DiffVsLastWeek'] = average_daily_sales[['Sales']].diff(periods=7)
```

```python
average_daily_sales[average_daily_sales.Open == 1].sort_values(by='DiffVsLastWeek')
```
