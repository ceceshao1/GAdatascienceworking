```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

# this is the standard import if you're using "formula notation" (similar to R)
import statsmodels.formula.api as smf
```

```python
# read data into a DataFrame
data = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv', index_col=0)
data.head()
```

# Checks for Linear Regression. It works best when: 
1. The data is normally distributed (but doesn’t have to be)

2. X’s are independent of each other (low multicollinearity)

3. X’s significantly explain y (have low p-values)





## Check 1. Distribution

#### Last time we plotted our data like this

## Check 2. Low Multicollinearity

```python
cmap = sns.diverging_palette(220, 10, as_cmap=True)

correlations = data[['TV', 'radio', 'newspaper']].corr()
print(correlations)
sns.heatmap(correlations, cmap=cmap, vmin=-1, vmax=1, annot=True)
```

### Student question: 
1. Do these variables have colinearity? 

**Answer**  Based on the heatmap There is a chance that radio and newspaper have some correlation, but the value of 0.35 is not necessarily high enough to suggest colinearity.

## Check 3: X’s significantly explain y (have low p-values)

#### Let's take a look again the the crude model

```python
lm = smf.ols(formula='sales ~ TV', data=data).fit()

#print the full summary
lm.summary()

#key numbers
#r-squared = 0.612 
#TV coefficient = 0.0475
# p-value < 0.05
```

### Student Model
#### Now do a full model with TV, Radio and Newspaper
syntax can be found here: http://statsmodels.sourceforge.net/devel/example_formulas.html

```python
#fit model
mod = smf.ols(formula='sales ~ TV + radio + newspaper', data=data)
res = mod.fit()

#print summary

print(res.summary())
```

#### 1. Which of the media buys were significantly associated with the sales? 

**Answer:** Both radio and TV media buys had p-values of less than 0.05 

#### 2. Controlling for all the other media buys, which media type had the largest association with sales?

**Answer:** radio, biggest coefficient

> putting variables in a regression model controls them, example = Intercept where you're controlling all the other variables (assigning them with 0 values)
